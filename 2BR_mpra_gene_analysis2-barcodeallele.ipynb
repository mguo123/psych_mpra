```r
save_pheatmap_pdf <- function(x, filename, width=7, height=7) {
   stopifnot(!missing(x))
   stopifnot(!missing(filename))
   pdf(filename, width=width, height=height)
   grid::grid.newpage()
   grid::grid.draw(x$gtable)
   dev.off()
}
```

```r
chrom_df = read.table('/Users/mguo123/Documents/pan_omics_psych/data/external/chrom_hg19.sizes',col.names=c('Chr','Position'))
chrom_df$Chr = factor(chrom_df$Chr, levels=c('chr1','chr2','chr3','chr4','chr5','chr6','chr7','chr8',
                                                 'chr9','chr10','chr11','chr12','chr13','chr14','chr15','chr16',
                                                 'chr17','chr18','chr19','chr20','chr21','chr22'))
chrom_df = chrom_df%>%drop_na()
chrom_df$disease = 'max'
```

```r
# make chromosome map of which loci are tested
diseases = c('schizo','depress','bipolar','anxiety','attent','personality','panic','traum','autism','ocd')
new_diseases = c('SCZ','MDD','BPD','GAD','ADHD','BLPD','PD','PTSD','ASD','OCD')

lib_df_sel = lib_df%>% select(c(diseases,'Chr','Position'))
lib_df_sel$Chr = factor(lib_df_sel$Chr, levels=c('chr1','chr2','chr3','chr4','chr5','chr6','chr7','chr8',
                                                 'chr9','chr10','chr11','chr12','chr13','chr14','chr15','chr16',
                                                 'chr17','chr18','chr19','chr20','chr21','chr22'))
lib_df_sel$num_dz = rowSums(lib_df_sel[,diseases])
lib_df_sel = lib_df_sel%>%
    rename("SCZ"='schizo')%>%
    rename("MDD"='depress')%>%
    rename("BPD"='bipolar')%>%
    rename("GAD"='anxiety')%>%
    rename("ADHD"='attent')%>%
    rename("BLPD"='personality')%>%
    rename("PD"='panic')%>%
    rename("PTSD"='traum')%>%
    rename("ASD"='autism')%>%
    rename("OCD"='ocd')#%>%
lib_df_sel_multi = lib_df_sel[lib_df_sel$num_dz>1,]
lib_df_sel_multi$disease = 'multiple'
lib_df_sel_single = lib_df_sel[lib_df_sel$num_dz==1 ,]
lib_df_sel_single$disease = apply(lib_df_sel_single[,new_diseases],1,function(x) new_diseases[x==1])
lib_df_sel = rbind(lib_df_sel_single,lib_df_sel_multi)
lib_df_sel = rbind(lib_df_sel[,c('Chr','Position','disease')],chrom_df)                                 
lib_df_sel$disease = factor(lib_df_sel$disease, 
                            levels = c('ADHD','ASD','BLPD','BPD','GAD','MDD','OCD','PD','PTSD','SCZ','multiple','max'))
lib_df_sel = unique(lib_df_sel)
lib_df_sel%>%
    group_by(disease)%>%tally()
ggplot(lib_df_sel, aes(x=Chr, y=Position, color=disease))+
                                  geom_point(alpha=.6)+ geom_jitter()+
                                  scale_color_viridis(option="viridis",discrete=TRUE)+#scale_color_manual(values=wes_palette(n=12, name="GrandBudapest"))+
                                  theme_classic()
ggsave('mpra_enrichment/library_chrom_map_wmax.pdf',width=12,height=5)
```

# per dz heatmaps

```r
diseases = c('anxiety', 'attent', 'autism', 'bipolar', 'depress', 'ocd', 'panic', 'personality', 'schizo', 'traum')
```

```r
factor_height = 50 # 50 rows --> 7 in
for (dz in diseases){
    dz_file = paste0('D_mpraanalyze_barcode_allelic/dz_snp_heatmaps/',dz,'.csv')
    dz_df = read.csv(dz_file)
    colnames(dz_df) = c('rowname','AST',colnames(dz_df)[3:length(colnames(dz_df))])
    dz_df = dz_df[,order(colnames(dz_df))]
    dz_df = column_to_rownames(dz_df, 'rowname')
    dz_df[dz_df< -2.5] = -2.5
    dz_df[dz_df> 2.5] = 2.5
    print(dz)
    print(dim(dz_df))
    if(dim(dz_df)[1]>1){
        new_height = 7*dim(dz_df)[1]/factor_height
        p = pheatmap(dz_df,fontsize_row=6,cluster_col=F,
                           color = colorRampPalette(rev(brewer.pal(n =8, name ="RdBu")))(255))
        save_pheatmap_pdf(p, paste0('D_mpraanalyze_barcode_allelic/dz_snp_heatmaps/',dz,'.pdf'),height=new_height,width=7)
    }

}
```

```r
mpra_df = read.csv('D_mpraanalyze_barcode_allelic/final_df.csv')
```

```r
# colnames(mpra_df)
mpra_df = mpra_df%>% select(fdr,pval,logFC)
mpra_df$pass_thres = mpra_df$fdr <0.05

mpra_df$neg_log10pval = -log10(mpra_df$pval)
mpra_df$logFC = -mpra_df$logFC
mpra_df = mpra_df%>%
    filter((logFC<2)& (logFC > -2))%>%
    filter(neg_log10pval<5)%>%
    filter(neg_log10pval>0.001)

ggplot(mpra_df, aes(x=logFC, y=neg_log10pval, color=pass_thres))+geom_point()+ scale_color_manual(breaks = c(FALSE, TRUE),
                        values=c("black", "#ff3333"))+ theme_classic()+theme(legend.position = "none")+ylab('-log(p)')
  
ggsave('D_mpraanalyze_barcode_allelic/overall_volcano.pdf',width=5,height=5)
```

```r
dz_stats_df = data.frame(disease=c('GAD','ADHD','ASD','BPD','MDD','OCD','PD','BLPD','SCZ','PTSD'),
          n_assayed_snps=c(179,142,54,400,424,116,61,39,1067,101),
          yes=c(56,46,10,105,117,36,15,12,322,30)) %>%
    arrange(disease)%>%
    mutate(disease=as.character(disease))%>%
    mutate(perc_sig = yes/n_assayed_snps)%>%
    mutate(no = n_assayed_snps - yes)
dz_stats_df
```

```r
dz_stats_df_long =  dz_stats_df%>%select(-n_assayed_snps)%>%pivot_longer(!disease, names_to='is_sig',values_to='num_snps')%>%
    mutate(is_sig=factor(is_sig, levels=c('yes','no')))
ggplot(dz_stats_df_long, aes(x=disease, y=num_snps, fill=is_sig)) + geom_bar(position='stack',stat='identity')+ scale_fill_manual(breaks = c(FALSE, TRUE),
                        values=c("#ff6666", "black"))+ theme_classic()+theme(legend.position = "none")+ylab('# SNPs')
ggsave('D_mpraanalyze_barcode_allelic/loci_stats.pdf',width=4,height=4)
```
