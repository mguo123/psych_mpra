```python
def process_egenestr(genestr):
    drug_dict = {'target_gene':[],'target_pertname':[], 'target_moa':[],'target_clinical_phase':[], 'target_indication':[]}
    egene_arr = genestr.split('|')
    
#     print(egene_arr)
    if len(egene_arr)>0:
        for gene in egene_arr:
            drug_targeted = repurposing_drugs_df_psych_unstacked[repurposing_drugs_df_psych_unstacked.gene==gene]
            drug_dict['target_gene']+=list(drug_targeted.gene)
            drug_dict['target_pertname']+=list(drug_targeted.pert_iname)
            drug_dict['target_moa']+=list(drug_targeted.moa)
            drug_dict['target_clinical_phase']+=list(drug_targeted.clinical_phase)
            drug_dict['target_indication']+=list(drug_targeted.indication)
    
    drug_dict_str = {}
    for key, arr in drug_dict.items():
        drug_dict_str[key]='|'.join(arr)
    return pd.Series(drug_dict_str)
    
```

## 4B. add gene to drug + direction info to the `lib_df_egenes_df_target` 

```python
def process_egenestr_cmap(genestr, gene_to_drug_dict_up,gene_to_drug_dict_dn):
    drug_dict = {'cmap_gene_up':[],'cmap_pertname_up':[], 'cmap_gene_dn':[],'cmap_pertname_dn':[]}
    egene_arr = genestr.split('|')
#     print(egene_arr)
    if len(egene_arr)>0:
        for gene in egene_arr:
            if gene in gene_to_drug_dict_up:
                drug_list = gene_to_drug_dict_up[gene]
                gene_list = [gene]*len(drug_list)
                drug_dict['cmap_gene_up']+=gene_list
                drug_dict['cmap_pertname_up']+=drug_list
            if gene in gene_to_drug_dict_dn:
                drug_list = gene_to_drug_dict_dn[gene]
                gene_list = [gene]*len(drug_list)
                drug_dict['cmap_gene_dn']+=gene_list
                drug_dict['cmap_pertname_dn']+=drug_list
    drug_dict_str = {}
    for key, arr in drug_dict.items():
        drug_dict_str[key]='|'.join(arr)
    return pd.Series(drug_dict_str)
    
```

```python
gene_up = []
for genestr in lib_df_egenes_df.unique_egenes_gtex:
    for gene in genestr.split('|'):
        if gene in genes_up_drug_dict1.keys():
            gene_up.append(gene)
print('number of gtex egenes that have a perturbagen that upregulates expression of gene:', len(set(gene_up)))

gene_dn = []
for genestr in lib_df_egenes_df.unique_egenes_gtex:
    for gene in genestr.split('|'):
        if gene in genes_dn_drug_dict1.keys():
            gene_dn.append(gene)
print('number of gtex egenes that have a perturbagen that downregulates expression of gene:', len(set(gene_dn)))
```

```python
lib_df_egenes_df_target_cmap = pd.concat([lib_df_egenes_df_target,
                                          lib_df_egenes_df_target.unique_egenes_gtex.apply(
                                              lambda x: process_egenestr_cmap(x, genes_up_drug_dict1,genes_dn_drug_dict1))],axis=1)
len(lib_df_egenes_df_target_cmap.cmap_pertname_up.unique()),len(lib_df_egenes_df_target_cmap.cmap_pertname_dn.unique())
```

```python
def get_drugs(dz_genes_df, cmap_gene_drug_dict, suffix=''):
    
    dz_genes_df['isin_'+suffix] = dz_genes_df.gene_sym.isin(cmap_gene_drug_dict.keys())
    dz_genes_df['drugs_'+suffix] = dz_genes_df.gene_sym.map(cmap_gene_drug_dict)
    dz_genes_df['num_drugs_'+suffix] = dz_genes['drugs_'+suffix].apply(len)
    return dz_genes_df
```

```python
def process_egene_file(dz_gene_row, dz):
    gene = dz_gene_row['gene_sym']
    
    
    # filter snp list for disease-matching and containing at least 1 of the  egene list 
    results = {'snps':'','logFC_str':'','logFC_mean':0.0}
    mpra_res_df_wfc_dz = mpra_res_df_wfc[mpra_res_df_wfc[dz]==1]
    mpra_res_df_wfc_dz_gene = mpra_res_df_wfc_dz[mpra_res_df_wfc_dz.unique_egenes_gtex.str.contains(gene)]
    if mpra_res_df_wfc_dz_gene.shape[0]>0:
        results['snps'] = '|'.join(mpra_res_df_wfc_dz_gene.Linked_SNP)
        results['logFC_str'] = '|'.join(mpra_res_df_wfc_dz_gene.logFC_str)
        results['logFC_mean'] = mpra_res_df_wfc_dz_gene.logFC_mean.mean()
    return pd.Series(results)
        
        
    
```
