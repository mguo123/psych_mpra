```python
tissue_to_sample = {
    'AST':['AST_1', 'AST_2'],
    'H9D0':['H9_D0_1A', 'H9_D0_1B', 'H9_D0_2'],
    'H9D2':['H9_D2_1', 'H9_D2_2','H9_D2_3', 'H9_D2_4'],
    'H9D4':['H4_D4_1','H9_D4_2', 'H9_D4_3A', 'H9_D4_3B'],
    'H9D10':[ 'H9_D10_1', 'H9_D10_3', 'H9_D10_4',],#'H9_D10_2',
    'SL':['SL_1','SL_2','SL_3A', 'SL_3B', 'SL_4'],
    'SLC':['SLC_1', 'SLC_2', 'SLC_3', 'SLC_4'],
    'HEK293':['HEK293_1',  'HEK293_2', 'HEK293_3']
    
}

map_tissue = {
    'AST':'AST',
    'H9D0':'ES',
    'H9D2':'N-D2',
    'H9D4':'N-D4',
    'H9D10':'N-D10',
    'SL':'A-NSC',
    'SLC':'P-NSC',
    'HEK293':'HEK293T',
    
}
```

```python
dna_df = pd.read_csv('../miseq_mg_081019/libpsych_plasmid_qc_701_501_count.csv',index_col=0).set_index('id')
dna_df
```

```python
get_rna('chr1', '99190949', 'SLC')
samples = tissue_to_sample['SLC']
rna_df[samples][rna_df.index.str.contains('chr1.99190949')]#.sum(axis=1)
```

```python
def get_rna(chrom, pos, tissue):
    index = chrom + '.'+ str(pos)
    samples = tissue_to_sample[tissue]
    rna_df_sel_sum = rna_df[samples][rna_df.index.str.contains(index)].sum(axis=1)
    rna_df_sel_sum.index.name = None
    return(rna_df_sel_sum)

def get_dna(chrom, pos):
    index = chrom + '.'+ str(pos)
    dna_df_sel_sum = dna_df[dna_df.index.str.contains(index)]['count']
    dna_df_sel_sum.index.name = None

    return(dna_df_sel_sum)
 
def make_plot(chrom,pos,tissue, show_fig=True,save_dir = save_dir):
    rna_c = get_rna(chrom,pos,tissue)
    dna_c = get_dna(chrom,pos)
    rna_dna = pd.DataFrame((rna_c+1).div(dna_c+1)).dropna()
    rna_dna.columns=['ratio']
    rna_dna = rna_dna[(np.abs(stats.zscore(rna_dna.ratio)) < 2.5)]#.all(axis=1)]
    rna_dna['is_ref'] = rna_dna.index.str.contains('Ref') 
    rna_dna['Ref/Alt'] = rna_dna.is_ref.apply(lambda x: 'Ref' if x else 'Alt')
    mean_ref_sign = rna_dna[rna_dna['Ref/Alt']=='Ref'].ratio.median()
    rna_dna['ratio'] = rna_dna.ratio/mean_ref_sign
    #     boxplot = rna_dna.boxplot(by='is_ref')
    plt.figure()
    if show_fig:
        display(rna_dna)
    ax = sns.swarmplot(x="Ref/Alt", y="ratio", data=rna_dna,color='black')
    ax = sns.boxplot(x="Ref/Alt", y="ratio", data=rna_dna, whis=np.inf)
#     ax = sns.violinplot(x="Ref/Alt", y="ratio", data=rna_dna)#,color='black')

    ax.set_title (chrom+'_'+ str(pos)+'_'+map_tissue[tissue])
    if save_dir is not None:
        plt.savefig(os.path.join(save_dir, chrom+'_'+ str(pos)+'_'+tissue+'.pdf'))
    if show_fig:
        plt.show()
    plt.close()
    return rna_dna

def make_overall_plot(chrom, pos, show_fig=True,save_dir = save_dir, 
                      tissues=tissue_to_sample.keys()):
    df_all = pd.DataFrame()
    for tissue in tissues:
        rna_c = get_rna(chrom,pos,tissue)
        dna_c = get_dna(chrom,pos)
        rna_dna = pd.DataFrame(rna_c.div(dna_c)).dropna()
        rna_dna.columns=['ratio']
        rna_dna = rna_dna[(np.abs(stats.zscore(rna_dna.ratio)) < 2.5)]#.all(axis=1)]
        rna_dna['is_ref'] = rna_dna.index.str.contains('Ref') 
        rna_dna['Ref/Alt'] = rna_dna.is_ref.apply(lambda x: 'Ref' if x else 'Alt')
        mean_ref_sign = rna_dna[rna_dna['Ref/Alt']=='Ref'].ratio.median()
        rna_dna['ratio'] = rna_dna.ratio/mean_ref_sign
        rna_dna['tissue'] = map_tissue[tissue]
        df_all = pd.concat([df_all, rna_dna])
    if df_all.shape[0]>0:
        plt.figure()
        ax = sns.boxplot(data=df_all, x="tissue", y="ratio",hue='Ref/Alt', 
                         palette=[sns.color_palette("Set2")[0]]+sns.color_palette("Set2")[3:],#["tomato","darkturquoise"],
                         dodge=True)
        ax.axhline(y=1,color='grey' ,dashes=(2,2))
        ax.set_title (chrom+'_'+ str(pos))
        plt.savefig(os.path.join(save_dir, chrom+'_'+ str(pos)+'_all.pdf'))
        if show_fig:
            plt.show()
        plt.close()

    return df_all
```
