## annotation stuff
- Snp names and such (DONE)
- diseases relevant, (DONE)

```python
lib_csv = '../../psych_lib_info.csv'
lib_df = pd.read_csv(lib_csv,index_col=0)
lib_df['rowname']= lib_df.Chr + '_' + lib_df.Position.map(str)

# lib_df = lib_df[lib_df.allele_type=='all_alt']
# lib_df = lib_df[['name','rowname','Linked_SNP', 'Chr', 'Position', 'Index_SNP','eGene','eQTL_tissue','Pval','source',
#                  'schizo', 'depress', 'bipolar', 'anxiety', 'attent', 'personality', 'panic', 'traum', 'autism', 'ocd','Consequence',
#                 ]]
diseases = ['schizo', 'depress', 'bipolar', 'anxiety', 'attent', 'personality', 'panic' ,'traum' ,'autism' ,'ocd']
diseases_abbr = ['SCZ','MDD','BPD','GAD','ADHD','BLPD','PD','PTSD','ASD','OCD']
def process_lib_dis(row):
    dis_str_arr = []
    for dz, dz_abbr in zip(diseases,diseases_abbr):
        if row[dz]==1:
            dis_str_arr.append(dz_abbr)
    return '|'.join(sorted(dis_str_arr))

lib_df['diseases'] = lib_df.apply(process_lib_dis,axis=1)
lib_df['num_dz'] = lib_df[diseases].sum(axis=1)
lib_df['Ref/Alt'] = lib_df.Ref + '/' + lib_df.Alt
print(lib_df.columns.values)
print(lib_df.shape)
print(lib_df.rowname.value_counts().shape)

lib_df_selcol = lib_df[['rowname','Linked_SNP','Chr','Position', 'Ref/Alt', # details for linked snp
                        'Index_SNP', 'ConsScore', 'ConsDetail', # details for index snp
                        'diseases','num_dz', 
                        'Pval','source',
                        'GeneName']].sort_values('Index_SNP')
lib_df_selcol.columns = ['rowname','Linked_SNP','Chr','Position', 'Ref/Alt', # details for linked snp
                        'Index_SNP', 'ConsScore', 'ConsDetail', # details for index snp
                        'diseases','num_dz', 
                        'gwas_pval','source', 
                        'gene_annon']
lib_df_selcol.drop_duplicates(inplace=True)
lib_df_selcol.reset_index(drop=True,inplace=True)
lib_df_selcol[:5]    
```

```python
lib_df_selcol.shape
```

## egenes from gtex & tissues egenes are active gtex: 'mpra_enrichment/lib_df_gtex_rna_filt.csv'


```python
lib_gtex_annotation = pd.read_csv('mpra_enrichment/lib_df_gtex_rna_filt.csv',index_col=0).fillna('')
print(lib_gtex_annotation.shape)
lib_gtex_annotation.columns
```

```python
lib_gtex_annotation = lib_gtex_annotation[['rowname', 'ngene_haploreg','egene_all', 'all_genes', 'all_gtex_tissues', 'unique_egenes_gtex']]
lib_gtex_annotation  = lib_gtex_annotation.drop_duplicates().reset_index(drop=True)
lib_gtex_annotation['egene_neurotissue_gtex'] = ''
lib_gtex_annotation['egene_tissuemap_gtex'] = ''
for idx, row in lib_gtex_annotation.iterrows():
    all_genes_arr = row['all_genes'].split('|')
    all_gtex_tissues = row['all_gtex_tissues'].split('|')
    tuple_gene_tissue_undupl = set(zip(all_genes_arr,all_gtex_tissues))
    if len(set(zip(all_genes_arr,all_gtex_tissues)))>0:
        tuple_gene_tissue_undupl = sorted(list(tuple_gene_tissue_undupl),key=lambda x: x[0])
        egene_neurotissue_gtex,egene_tissuemap_gtex = map(list,zip(*tuple_gene_tissue_undupl))
        lib_gtex_annotation.at[idx, 'egene_neurotissue_gtex'] = '|'.join(egene_neurotissue_gtex)
        lib_gtex_annotation.at[idx, 'egene_tissuemap_gtex'] = '|'.join(egene_tissuemap_gtex)
```

```python
lib_gtex_annotation = lib_gtex_annotation.drop(['all_genes', 'all_gtex_tissues'],axis=1)
lib_gtex_annotation.columns = ['rowname','ngene_haploreg','egene_alltissue_gtex','egene_neurotissue_gtex',
                               'egene_neurotissue_mapgene_gtex','egene_neurotissue_maptissue_gtex']
print(lib_gtex_annotation.shape)
```

```python
lib_df_selcol_egene = lib_df_selcol.merge(lib_gtex_annotation,how='left',on='rowname')
lib_df_selcol_egene[:5]
```

## mpra results
- mpra res, mpra logfc change and pval, tissues effected
- egenes that have chromatin linkages - only exists for mpra hits b/c tissue must match mpra 


```python
mpra_res_df = pd.read_csv('D_mpraanalyze_barcode_allelic/final_df_filt.csv',index_col=0).sort_values('tissue')
mpra_res_df = mpra_res_df.groupby('rowname').agg({
    'tissue':'|'.join, 
    'pval':[lambda arr: '|'.join("%.3f"%x for x in arr),'mean'],
    'logFC':[lambda arr: '|'.join("%.3f"%x for x in arr),'mean', ]}).reset_index()
mpra_res_df.columns= ['rowname','mpra_tissue','mpra_pval_str','mpra_pval_mean','mpra_logfc_str','mpra_logfc_mean']
mpra_res_df['bool_mpra_sig'] = True
print(mpra_res_df.shape, mpra_res_df.columns)
mpra_res_df[:5]
```

```python
mpra_cgene_egene_df = pd.read_csv('D_mpraanalyze_barcode_allelic/snp_all_results_cgene_egene.csv',index_col=0)
mpra_cgene_egene_df = mpra_cgene_egene_df[['rowname','cgene_egene_agree','cgene_egene_agree_num']]
mpra_cgene_egene_df['bool_cgene_egene_agree'] = True
print(mpra_cgene_egene_df.shape, mpra_cgene_egene_df.columns)
mpra_cgene_egene_df[:5]
```

```python
lib_df_selcol_egene_atachichip_motif_drug_as_mpra = lib_df_selcol_egene_atachichip_motif_drug_as.merge(mpra_res_df,how='left',on='rowname')
lib_df_selcol_egene_atachichip_motif_drug_as_mpra.mpra_tissue.fillna('',inplace=True)
lib_df_selcol_egene_atachichip_motif_drug_as_mpra.mpra_pval_str.fillna('',inplace=True)
lib_df_selcol_egene_atachichip_motif_drug_as_mpra.mpra_pval_mean.fillna(0.0,inplace=True)
lib_df_selcol_egene_atachichip_motif_drug_as_mpra.mpra_logfc_str.fillna('',inplace=True)
lib_df_selcol_egene_atachichip_motif_drug_as_mpra.mpra_logfc_mean.fillna(0.0,inplace=True)
lib_df_selcol_egene_atachichip_motif_drug_as_mpra.bool_mpra_sig.fillna(False,inplace=True)

lib_df_selcol_egene_atachichip_motif_drug_as_mpra = lib_df_selcol_egene_atachichip_motif_drug_as_mpra.merge(mpra_cgene_egene_df,how='left',on='rowname')
lib_df_selcol_egene_atachichip_motif_drug_as_mpra.cgene_egene_agree.fillna('',inplace=True)
lib_df_selcol_egene_atachichip_motif_drug_as_mpra.cgene_egene_agree_num.fillna(0,inplace=True)
lib_df_selcol_egene_atachichip_motif_drug_as_mpra.bool_cgene_egene_agree.fillna(False,inplace=True)
```

## SO EXTRA STUFF SO NGENE-CGENE-EGENE AGREEMENT?
pie chart of SNPs making ngene( the annotated gene)
-  doesn't agree with cgene or egene
- agrees with Cgene but not egene
- agrees with ngene but not egene
- agrees with all

```python
cgene_df = pd.read_csv('D_mpraanalyze_barcode_allelic/snp_all_results_cgene_egene_tissue_gtex.csv')
print(cgene_df.shape,cgene_df.columns)

rowname_to_cgene_dict = dict(zip(cgene_df.rowname, pd.Series(cgene_df['cgenes_anchor'].str.split('|') + cgene_df['cgenes_loop'].str.split('|'))))
```

```python
count_gene_annon = {'ngene_only':0, 'ngene_cgene':0, 'ngene_egene':0, 'all_agree':0, 'no_annotation':0}
print(lib_df_annon_mprasig_FINAL.shape,lib_df_annon_mprasig_FINAL.columns)
for idx, row in lib_df_annon_mprasig_FINAL.iterrows():
        ngene = row['gene_annon']
    
    
        cgenes=rowname_to_cgene_dict.get(row['rowname'],'')
        egenes = row['egene_neurotissue_gtex'].split('|')
        if type(ngene)==float:
            count_gene_annon['no_annotation']+=1
        else:
            if ngene in cgenes:
                if ngene in egenes:
                    count_gene_annon['all_agree']+=1
                else:
                    count_gene_annon['ngene_cgene']+=1
            else:
                if ngene in egenes:
                    count_gene_annon['ngene_egene']+=1
                else:
                    count_gene_annon['ngene_only']+=1
count_gene_annon                   
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
pd.Series(count_gene_annon).plot.pie()
```

```python
pie, ax = plt.subplots(figsize=[10,6])
data = pd.Series(count_gene_annon)
labels = data.keys()
plt.pie(x=data, autopct="%.1f%%", explode=[0.05]*len(data), labels=labels, pctdistance=0.5)
plt.title("Annotation Agreement between different modalities", fontsize=14);
pie.savefig("figure_tables/count_gene_annon.svg")
```
