For the subset of LD (linkage disequilibrium)-independent SNPs with P < 1×10−6 in the meta-analysis, we next obtained results from deCODE genetics (1,513 cases and 66,236 controls of European ancestry). We define LD-independent SNPs as those with low LD (r2 < 0.1) to a more significantly associated SNP within a 500 kb window. Given high LD in the extended MHC region spans ~8 Mb, we conservatively include only a single MHC SNP to represent this locus. The deCODE data were then combined with those from the primary GWAS to give a dataset of 36,989 cases and 113,075 controls.
```r
dist = 500000
lib_df_sel = lib_df%>%arrange(Chr, Position)%>%
    mutate(pos_round1000 = round_any(Position, dist) )%>%
    group_by(Chr,pos_round1000)%>%
    dplyr::mutate(Linked_SNP_test=first(Linked_SNP))%>%
    ungroup()%>%
    dplyr::select(rowname, Linked_SNP,Linked_SNP_test, Chr, Position,pos_round1000,diseases,dz_new)
tested_SNPS = unique(lib_df_sel$Linked_SNP_test)
length(tested_SNPS)
```

```r
# Generates a data frame of pairwise linkage disequilibrium statistics. 
# Input is a list of between 2 to 1000 variants. 
# Desired output can be based on estimates of R2 or D’.
# note neeed to do per chromosome
for (chr in sort(unique(lib_df$Chr))){
    print(chr)
    tested_SNPS = lib_df$Linked_SNP[lib_df$Chr==chr]
    print(length(tested_SNPS))
    ld_matrix = LDmatrix(snps = tested_SNPS, 
         pop = "ALL", r2d = "r2", 
         token = token,
          file = paste0('ld_block/ld_mat_',chr,'.csv')          
        )
}

```

```r
thres = 0.1
```

```r
rsid_to_groupid_all = data.frame()
for (chr in sort(unique(lib_df$Chr))){
    # for each chromosome, get ld_mat
    file = paste0('ld_block/ld_mat_',chr,'.csv')  
    ld_mat = read.table(file, sep='\t',header=TRUE,row.names=1)
    # binarize ld based on thres
    ld_mat_bin = ld_mat
    ld_mat_bin[ld_mat_bin>thres] = 1
    ld_mat_bin[ld_mat_bin<=thres] = 0
    # make a df of Linked_SNP, group_id, and chrom from
    rsid_to_groupid = data.frame(sapply(colnames(ld_mat_bin), 
                  function(rsid) 
                      sort(rownames(ld_mat_bin)[ld_mat_bin[[rsid]]==1])[1]
                  )
          )
    rsid_to_groupid = rownames_to_column(rsid_to_groupid, 'Linked_SNP')
    colnames(rsid_to_groupid) = c('Linked_SNP','group_id') 
    rsid_to_groupid$group_id = as.character(rsid_to_groupid$group_id)
    rsid_to_groupid$chr_ld = chr
    rsid_to_groupid_all = rbind(rsid_to_groupid_all, rsid_to_groupid)
}
```
