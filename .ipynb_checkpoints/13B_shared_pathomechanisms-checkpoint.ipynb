```python
import pandas as pd
import glob,os,re
```

```python
save_dir = 'ukbiobank_phewas/shared_pathomechanism'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
mpra_res_df = pd.read_csv('figure_tables/mpra_res_df_UPDATED_ABC.csv',index_col=0).fillna('')
mpra_res_df.columns
```

```python
mpra_res_df_filt = mpra_res_df[mpra_res_df.ukbb_phenotype!='']
print(mpra_res_df_filt.shape)
mpra_res_df_filt = mpra_res_df_filt[mpra_res_df_filt.egenes_all_sources!='']
print(mpra_res_df_filt.shape)

```

```python
all_phenotypes = set()
for phe in mpra_res_df_filt.ukbb_phenotype:
    all_phenotypes = all_phenotypes.union(set(phe.split('|')))
all_phenotypes = sorted(all_phenotypes)
print(len(all_phenotypes))
print(sorted(all_phenotypes))
```

```python
# testing get names for saving
for phe in all_phenotypes:
    print(phe, '\t', '_'.join(re.split('[- \/]',phe)))
```

```python
phe_to_egenes = {}
for phe in all_phenotypes:
    # get snps
    print('+++++++++')
    phe_df = mpra_res_df_filt[mpra_res_df_filt.ukbb_phenotype.str.contains(phe)]
    phe_str =  '_'.join(re.split('[- \/]',phe))
    print(phe, phe_str)
    print(phe_df.shape[0],'num snps with phe')
    
    # get snp, egene disease linkage
    phe_df_sel = phe_df[['Linked_SNP','egenes_all_sources','diseases']]
    phe_df_sel = (phe_df_sel
    .assign(egenes_all_sources=phe_df_sel.egenes_all_sources.str.split('|'))
    .explode('egenes_all_sources')
    .assign(diseases=phe_df_sel.diseases.str.split('|'))
    .explode('diseases')
    )

    phe_genes = sorted(phe_df_sel.egenes_all_sources.unique())
    print(len(phe_genes),'num egenes with phe')
    print(phe_genes)
    phe_to_egenes[phe]=phe_genes
    with open(os.path.join(save_dir, 'genes_'+phe_str+'.txt'),'w') as f:
        for g in phe_genes:
            f.write(g+'\n')
    
    phe_dz = phe_df_sel.diseases.unique()
    print(phe_dz, len(phe_dz),'num diseases with phe')
    phe_df_sel.to_csv(os.path.join(save_dir,'phe_df_'+ phe_str+'.csv'))
    
    phe_df_by_gene = (phe_df_sel
     .groupby('egenes_all_sources')
     .agg({
         'diseases':lambda arr: '|'.join(sorted(set(arr))),
         'Linked_SNP':[lambda arr: '|'.join(sorted(set(arr))),'count']

     })
     .reset_index()
     .set_axis(['egene','diseases','Linked_SNP','num_snps'],axis=1,inplace=False)
    )
    phe_df_by_gene.to_csv(os.path.join(save_dir,'phe_df_gene_'+ phe_str+'.csv'))
    
    # for cytoscape
    edge_df = pd.concat([
        phe_df_sel[['Linked_SNP', 'egenes_all_sources']].set_axis(['node1','node2'],inplace=False, axis=1)
#         phe_df_sel[['diseases', 'egenes_all_sources']].set_axis(['node1','node2'],inplace=False, axis=1)    
    ]).drop_duplicates()
    node_df = phe_df_sel.melt().set_axis(['node_type','node'],inplace=False, axis=1).drop_duplicates()
    print(edge_df.shape[0], 'num edges')
    print(node_df.shape[0], 'num nodes')
    edge_df.to_csv(os.path.join(save_dir,'network_'+ phe_str+'_edge.csv'))
    node_df.to_csv(os.path.join(save_dir,'network_'+ phe_str+'_node.csv'))


#     phe_genes = sorted(set('|'.join(phe_df.egenes_all_sources).split('|')))
```

# how to make networks

1. pick a phenotype - good phenotypes are phenotypes with lots of edges in string (step 2) have lots of diseases, and a good # of genes)
2. make snp-gene network:
    - a. drag <network_'+ phe_str+'_edge.csv> to the network section, default source/target
    - b. drag <network_'+ phe_str+'_node.csv> to the node section, select node as the key
    - c. drag <phe_df_gene_'+ phe_str+'.csv'> to the node section, select egene as the key
    - d. hit the pathomechanisms style option
    - e. (opt) hit Layout>preferred layout and save a non string-field network image as svg
3. get string network - by uploading genes for phe to string, then select export to Cytoscape (make sure cytoscape is open and accepting network connections) --> refer to this as string_network_phe (don't change the name of the string network until at the very end)
4. merge string_network_phe and snp-gene network (Tools>Merge>Merge Networks)
    - a. make sure name and shared_name from snp-gene network both map to "display name in string_network
    - b. hit the pathomechanisms style option make sure style is good
    - c. hit Layout>preferred layout and save a string-field network image as svg (may require some manual dragging)

# go term networks


```python
go_terms_df = pd.read_excel('D_mpraanalyze_barcode_allelic/egene_gtex_psychencode/ggo_bp.xlsx',sheet_name='selected_go_terms').fillna('')
go_terms_df
```

```python
for go_term in go_terms_df.columns:
    print('+++++++++')
    phe_str = '_'.join(re.split('[- \/]',go_term))
    print(go_term, phe_str)
    
    phe_genes = go_terms_df[go_term][go_terms_df[go_term]!=''].sort_values().unique()
    print(len(phe_genes),'num egenes with phe')
    print(phe_genes)
    with open(os.path.join(save_dir, 'genes_'+phe_str+'.txt'),'w') as f:
        for g in phe_genes:
            f.write(g+'\n')   

    # get snps that match the egenes
    phe_df = mpra_res_df[mpra_res_df.egenes_all_sources.apply(lambda x: any(map(x.__contains__, phe_genes)))]
    print(phe_df.shape[0],'num snps with phe')

    # get snp, egene disease linkage
    phe_df_sel = phe_df[['Linked_SNP','egenes_all_sources','diseases']]
    phe_df_sel = (phe_df_sel
    .assign(egenes_all_sources=phe_df_sel.egenes_all_sources.str.split('|'))
    .explode('egenes_all_sources')
    .assign(diseases=phe_df_sel.diseases.str.split('|'))
    .explode('diseases')
    )
    phe_dz = phe_df_sel.diseases.unique()
    print(phe_dz, len(phe_dz),'num diseases with phe')
    phe_df_sel.to_csv(os.path.join(save_dir,'phe_df_'+ phe_str+'.csv'))
    
    phe_df_by_gene = (phe_df_sel
     .groupby('egenes_all_sources')
     .agg({
         'diseases':lambda arr: '|'.join(sorted(set(arr))),
         'Linked_SNP':[lambda arr: '|'.join(sorted(set(arr))),'count']

     })
     .reset_index()
     .set_axis(['egene','diseases','Linked_SNP','num_snps'],axis=1,inplace=False)
    )
    phe_df_by_gene.to_csv(os.path.join(save_dir,'phe_df_gene_'+ phe_str+'.csv'))

    # for cytoscape
    edge_df = pd.concat([
        phe_df_sel[['Linked_SNP', 'egenes_all_sources']].set_axis(['node1','node2'],inplace=False, axis=1)
#         phe_df_sel[['diseases', 'egenes_all_sources']].set_axis(['node1','node2'],inplace=False, axis=1)    
    ]).drop_duplicates()
    edge_df = edge_df[edge_df.node2.isin(phe_genes)]
    node_df = phe_df_sel[['Linked_SNP', 'egenes_all_sources']].melt().set_axis(['node_type','node'],inplace=False, axis=1).drop_duplicates()
    node_df = node_df[(node_df.node.isin(phe_genes))|(node_df.node_type=='Linked_SNP')]
    print(edge_df.shape[0], 'num edges')
    print(node_df.shape[0], 'num nodes')
    edge_df.to_csv(os.path.join(save_dir,'network_'+ phe_str+'_edge.csv'))
    node_df.to_csv(os.path.join(save_dir,'network_'+ phe_str+'_node.csv'))
```
