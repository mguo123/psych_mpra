```python
import pandas as pd
import glob,os
from collections import defaultdict
from matplotlib import pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
```

```python
dz_to_abbr_dict = {
    'schizo':'SCZ',
    'bipolar':'BPD',
    'personality':'BPLD',
    'traum':'PTSD',
    'anxiety':'GAD',
#     'depress':'MDD',
    'depress_updatedhoward':'MDD',
    'ocd':'OCD',
    'panic':'PD',
    'autism':'ASD',
    'attent':'ADHD',
}
```

```python
save_dir = 'common_genes'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
def search_genes(key, col_to_search, df_to_search, cols_to_return):
    df_filt = df_to_search[df_to_search[col_to_search].str.contains(key)]
    results_dict = {}
    for col in cols_to_return:
        try:
            col_res = '|'.join([str(x) for x in sorted(set(df_filt[col]))])
        except:
            col_res = ''
        results_dict[col] = col_res
    return pd.Series(results_dict)
```

```python
lib_df = pd.read_csv('figure_tables/lib_df_annon_FINAL.csv',index_col=0).fillna('')
print(lib_df.shape)
lib_df.columns
```

```python
# mpra_df = pd.read_excel('figure_tables/lib_df_annon_mprasig_FINAL_v020521.xlsx',index_col=0).fillna('')
mpra_df = pd.read_csv('figure_tables/lib_df_annon_mprasig_FINAL_updatedMDD.csv',index_col=0).fillna('')
mpra_df.columns
```

```python
mpra_df_sel = mpra_df[['rowname','Linked_SNP', 'Chr', 'Position', 'Ref/Alt', 'Index_SNP','gwas_pval','mpra_tissue', 'mpra_pval_str',
       'mpra_pval_mean', 'mpra_logfc_str', 'mpra_logfc_mean', 
        'egene_neurotissue_gtex']]
mpra_df_sel
```

```python
egene_lists = sorted(glob.glob('D_mpraanalyze_barcode_allelic/egene_gtex/*txt'))
egene_lists
```

```python
genes_df_all = pd.DataFrame(columns=['gene','tissue'])
for genefile in egene_lists:
    tissue = os.path.basename(genefile).split('.')[0]
    if tissue in dz_to_abbr_dict:
        genelist = pd.read_table(genefile,sep='\t',header=None).iloc[:,0]
        print(tissue, len(genelist))
        genes_df = pd.DataFrame(pd.Series(genelist,name='gene'))
        genes_df['tissue'] = dz_to_abbr_dict[tissue]
        genes_df_all = pd.concat([genes_df_all,genes_df ])
```

```python
genes_dz_count = genes_df_all.groupby('gene').agg({'tissue':['count',lambda x: '|'.join(sorted(x))]}).reset_index()
genes_dz_count.columns = ['gene','count_dz','diseases']
genes_dz_count.sort_values('count_dz',ascending=False,inplace=True)
```

```python
gene_info_res = genes_dz_count.gene.apply(lambda x: search_genes(key=x, col_to_search='egene_neurotissue_gtex', 
             df_to_search = mpra_df, cols_to_return = ['rowname', 'Linked_SNP',
                                                       'Chr', 'Position', 'Ref/Alt',
                                                       'Index_SNP','gwas_pval',
                                                      'mpra_pval_mean',  'mpra_logfc_mean']))

```

```python
from collections import defaultdict
genes_dz_df_bin = pd.DataFrame(columns =list(dz_to_abbr_dict.values()))
for idx, row in genes_dz_count.iterrows():
    diseases = row['diseases'].split('|')
    for dz in diseases:
        genes_dz_df_bin.at[row['gene'],dz] = 1
```

```python
genes_dz_df_bin = genes_dz_df_bin.fillna(0)
genes_dz_df_bin.to_csv('common_genes/genes_dz_df_bin.csv')
```

```python
gene_info_res
```

```python
common_gene_df = pd.concat([genes_dz_count,gene_info_res],axis=1)
common_gene_df.shape
```
